```python
import pyarrow as pa
import pyarrow.compute as pc
import pyarrow.parquet as pq
```

```python
days = pa.array([1,2,3,4,5], type=pa.int8())
month = pa.array(['Jan','Feb','Mar','Jan','Feb'], type=pa.string())
year = pa.array([2020,2021,2022,2023,2024], type=pa.int16())

birth_table = pa.table([days,month,year], names=["Days", "Month", "Year"])
```

```python
birth_table
```

```python
pc.value_counts(birth_table['Month'])
```
